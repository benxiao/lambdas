```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("IMG_0407.jpg")
```

```python
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
sub = gray[400:1000, 0:750]
```

```python
plt.imshow(gray)
```

```python
sub_result = sub.copy()
```

```python
result = (sub_result.flatten() < 120).reshape(sub.shape) * 255
```

```python
plt.imshow(result)
```

```python
cv2.imwrite("gray.jpg", result)
```
